```python
import numpy as np
import pandas as pd
df = pd.read_csv('./healthcare-dataset-stroke-data.csv')
df
```

```python
df.describe()
```

```python
df['bmi'] = df['bmi'].replace(np.NaN, 28.893237)
df.describe()
```

```python
df['gender'] = df['gender'].replace('Male', 1)
df['gender'] = df['gender'].replace('Other', 0.5)
df['gender'] = df['gender'].replace('Female', 0)
df['ever_married'] = df['ever_married'].replace('Yes',1)
df['ever_married'] = df['ever_married'].replace('No',0)
df['Residence_type'] = df['Residence_type'].replace('Urban',1)
df['Residence_type'] = df['Residence_type'].replace('Rural',0)
df
```

```python
df.drop(labels='id', axis=1)
```

```python
categoricial_features = ['work_type','smoking_status']
df = pd.get_dummies(df, columns=categoricial_features)
df
```

```python
numeric_features = ['age', 'avg_glucose_level', 'bmi']
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(df[numeric_features])
df[numeric_features] = ss.transform(df[numeric_features])
df
```

```python
stroke = df[df['stroke'] == 1]
```

```python
df = df.drop(labels='id', axis=1)
df
```

```python
from sklearn.model_selection import train_test_split
df1 = df.drop(labels='stroke', axis=1)

X= df1  
y=df['stroke']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state= 88)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
model = clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print('Accuracy Score :', accuracy_score(y_test,y_pred))
print('Confusion Matrix :', confusion_matrix(y_test,y_pred))
print('Classification Report :', classification_report(y_test,y_pred))

performance = pd.DataFrame(columns=['Model Name','Accuracy'])
performance = performance.append({'Model Name': 'Random Forest', 'Accuracy': accuracy_score(y_test,y_pred)*100}, ignore_index=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=9)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print('Accuracy Score :', accuracy_score(y_test,y_pred))
print('Confusion Matrix :', confusion_matrix(y_test,y_pred))
print('Classification Report :', classification_report(y_test,y_pred))

performance = performance.append({'Model Name': 'KNN', 'Accuracy': accuracy_score(y_test,y_pred)*100}, ignore_index=True)
```

```python
from sklearn import svm
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print('Accuracy Score :', accuracy_score(y_test,y_pred))
print('Confusion Matrix :', confusion_matrix(y_test,y_pred))
print('Classification Report :', classification_report(y_test,y_pred))

performance = performance.append({'Model Name': 'SVM', 'Accuracy': accuracy_score(y_test,y_pred)*100}, ignore_index=True)
```

```python
from sklearn.linear_model import LogisticRegression 
clf = LogisticRegression(random_state = 95) 
model = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print('Accuracy Score :', accuracy_score(y_test,y_pred))
print('Confusion Matrix :', confusion_matrix(y_test,y_pred))
print('Classification Report :', classification_report(y_test,y_pred))

performance = performance.append({'Model Name': 'Logistic Regression', 'Accuracy': accuracy_score(y_test,y_pred)*100}, ignore_index=True)
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print('Accuracy Score :', accuracy_score(y_test,y_pred))
print('Confusion Matrix :', confusion_matrix(y_test,y_pred))
print('Classification Report :', classification_report(y_test,y_pred))

performance = performance.append({'Model Name': 'Naive Bayes', 'Accuracy': accuracy_score(y_test,y_pred)*100}, ignore_index=True)
```
